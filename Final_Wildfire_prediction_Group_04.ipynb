```python
#importing required libraries
from tifffile import imread
from skimage.transform import resize
import numpy as np
from glob import glob
from concurrent.futures import ThreadPoolExecutor
from datetime import datetime
import os
import re
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plot
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score, precision_score, recall_score, f1_score
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
import folium
import rasterio
from matplotlib import cm
from PIL import Image


# Setting Dataset Parameters
base_path = "/content/drive/MyDrive/AI_project/2021/"
target_shape = (64, 64)
num_workers = 8
timesteps = 5


# Retrieve All .tif Files Recursively
tif_files = sorted(glob(os.path.join(base_path, '**', '*.tif'), recursive=True))


# Extract date from filename using regex to enable chronological ordering
def extracting_time_from_filename(filename):
   match = re.search(r'(\d{4})[-_](\d{2})[-_](\d{2})', filename)
   if match:
       date_str = f"{match.group(1)}-{match.group(2)}-{match.group(3)}"
       try:
           datetime.strptime(date_str, "%Y-%m-%d")
           return date_str
       except ValueError:
           print(f"Invalid date: {date_str} in {filename}")
           return None
   return None


# Load and preprocess each .tif image file
def loading_tif(f):
   try:
       img = imread(f)


       # Skip completely flat or malformed files before casting
       if img.ndim == 1 or img.size < 64 * 64:
           print(f"Skipping corrupted or flat .tif: {f}")
           return None

       # Convert to float and normalize
       img = img.astype(np.float32)
       img = img[0] if img.ndim == 3 else img
       img = np.nan_to_num(img)
       img_resized = resize(img, target_shape, preserve_range=True, anti_aliasing=True)
       img_scaled = (img_resized - np.min(img_resized)) / (np.max(img_resized) - np.min(img_resized) + 1e-8)


       time_key = extracting_time_from_filename(f)
       return (time_key, img_scaled) if time_key else None


   except Exception as e:
       print(f"Error loading {f}: {e}")
       return None


       img = img[0] if img.ndim == 3 else img
       img = np.nan_to_num(img)
       img_resized = resize(img, target_shape, preserve_range=True, anti_aliasing=True)
       img_scaled = (img_resized - np.min(img_resized)) / (np.max(img_resized) - np.min(img_resized) + 1e-8)


       time_key = extracting_time_from_filename(f)
       if time_key is None:
           return None


       return time_key, img_scaled


   except Exception as e:
       print(f"Error: {f} — {e}")
       return None
       img = img[0] if img.ndim == 3 else img
       img = np.nan_to_num(img)
       img_resized = resize(img, target_shape, preserve_range=True, anti_aliasing=True)
       img_scaled = (img_resized - np.min(img_resized)) / (np.max(img_resized) - np.min(img_resized) + 1e-8)
       time_key = extracting_time_from_filename(f)
       if time_key is None:
           return None
       return time_key, img_scaled
   except Exception as e:
       print(f"Error: {f} — {e}")
       return None
# Load and process all images using multithreading
with ThreadPoolExecutor(max_workers=num_workers) as ex:
   processed = list(ex.map(loading_tif, tif_files))
# Filter out failed loads and sort by extracted date
processed = [p for p in processed if p and p[0] is not None]
print(f"Processed {len(processed)} valid .tif files out of {len(tif_files)}")
print(f"Processed {len(processed)} valid .tif files out of {len(tif_files)}")
print(f"Processed {len(processed)} valid .tif files out of {len(tif_files)}")


if len(processed) == 0:
   raise ValueError("No valid .tif files were processed. Check for compression issues or install imagecodecs.")


processed.sort(key=lambda x: x[0])
dates, image_sequence = zip(*processed)
image_sequence = np.array(image_sequence)[..., np.newaxis]
print("Shape of Image Sequence:", image_sequence.shape)
print("Date range:", dates[0], "to", dates[-1])
print("Shape of Image Sequence:", image_sequence.shape)
print("Date range:", dates[0], "to", dates[-1])
print("Shape of Image Sequence:", image_sequence.shape)
print("Date range:", dates[0], "to", dates[-1])


# Prepare training data using sliding window approach
X, Y = [], []
for i in range(len(image_sequence) - timesteps):
   X.append(image_sequence[i:i + timesteps])
   Y.append(image_sequence[i + timesteps])
X = np.array(X)
Y = np.array(Y)


# Define CNN-LSTM model architecture
def cnn_lstm_model(input_shape, timesteps):
   model = models.Sequential([
       layers.TimeDistributed(layers.Conv2D(16, (3, 3), activation='relu', padding='same'),
                              input_shape=(timesteps,) + input_shape),
       layers.TimeDistributed(layers.MaxPooling2D((2, 2))),
       layers.TimeDistributed(layers.Conv2D(32, (3, 3), activation='relu', padding='same')),
       layers.TimeDistributed(layers.MaxPooling2D((2, 2))),
       layers.TimeDistributed(layers.Flatten()),
       layers.LSTM(128),
       layers.Dense(np.prod(input_shape), activation='sigmoid'),
       layers.Reshape(input_shape)
   ])
   model.compile(optimizer='adam', loss='mse')
   return model
# Build and train the model
model = cnn_lstm_model(input_shape=(64, 64, 1), timesteps=timesteps)
model.summary()


history = model.fit(X, Y, batch_size=8, epochs=10, validation_split=0.1, verbose=1)


#Plot loss curves
plot.figure(figsize=(8, 5))
plot.plot(history.history['loss'], label='Training Loss', marker='o')
plot.plot(history.history['val_loss'], label='Validation Loss', marker='x')
plot.title("Model Loss Over Epochs")
plot.xlabel("Epoch")
plot.ylabel("Loss (MSE)")
plot.legend()
plot.grid(True)
plot.tight_layout()
plot.show()


# Predict the next frame based on the latest input sequence
predicted_next = model.predict(X[-1][np.newaxis, ...])


# Visualize prediction vs ground truth
plot.figure(figsize=(10, 4))
plot.subplot(1, 2, 1)
plot.imshow(Y[-1].squeeze(), cmap='hot')
plot.title("Ground Truth")
plot.subplot(1, 2, 2)
plot.imshow(predicted_next.squeeze(), cmap='hot')
plot.title("Predicted")
plot.tight_layout()
plot.show()


#Histogram comparison of predicted and actual pixel values
def plot_pixel_distribution(y_true, y_pred):
   plot.figure(figsize=(8, 5))
   plot.hist(y_true.flatten(), bins=50, alpha=0.5, label="Ground Truth")
   plot.hist(y_pred.flatten(), bins=50, alpha=0.5, label="Prediction")
   plot.title("Pixel Intensity Distribution")
   plot.xlabel("Pixel Value")
   plot.ylabel("Count")
   plot.legend()
   plot.grid(True)
   plot.tight_layout()
   plot.show()


plot_pixel_distribution(Y[-1], predicted_next)


# Evaluate model on last few frames
num_eval_frames = 5
X_seq = X[-num_eval_frames:]
Y_seq = Y[-num_eval_frames:]
y_pred_seq = model.predict(X_seq)


def evaluation_of_sequence(y_true_seq, y_pred_seq, threshold=0.5):
   mse_list, mae_list, psnr_list, ssim_list, acc_list  = [], [], [], [], []
   for t in range(len(y_true_seq)):
       y_true = y_true_seq[t]
       y_pred = y_pred_seq[t]
       mse_list.append(mean_squared_error(y_true.flatten(), y_pred.flatten()))
       mae_list.append(mean_absolute_error(y_true.flatten(), y_pred.flatten()))
       psnr_list.append(psnr(y_true.squeeze(), y_pred.squeeze(), data_range=1.0))
       ssim_list.append(ssim(y_true.squeeze(), y_pred.squeeze(), data_range=1.0))
       # Compute binary accuracy as an additional metric
       y_true_bin = (y_true > 0.5).astype(int).flatten()
       y_pred_bin = (y_pred > 0.5).astype(int).flatten()
       acc_list.append(accuracy_score(y_true_bin, y_pred_bin))
       plot.figure(figsize=(10, 3))
       plot.subplot(1, 2, 1)
       plot.imshow(y_true.squeeze(), cmap='hot')
       plot.title(f"Ground Truth Frame {t}")
       plot.subplot(1, 2, 2)
       plot.imshow(y_pred.squeeze(), cmap='hot')
       plot.title(f"Predicted Frame {t}")
       plot.tight_layout()
       plot.show()
   print("Evaluation Over Sequence")
   print(f"Mean Square Error(MAE): {np.mean(mse_list):.4f}")
   print(f"Mean Absolute Error(MAE): {np.mean(mae_list):.4f}")
   print(f"Peak Signal Noise Ratio(PSNR): {np.mean(psnr_list):.2f}")
   print(f"Structure Similarity Index Measure : {np.mean(ssim_list):.4f}")
   print(f"Accuracy : {np.mean(acc_list):.4f}")


evaluation_of_sequence(Y_seq, y_pred_seq)
```

```python
import folium
import numpy as np
from matplotlib import cm
import branca.colormap as bcm


def overlay_fire_colormap_points_with_legend(prediction,
                                             bbox=[[24.0, -125.0], [50.0, -66.5]],
                                             map_center=[39.5, -98.35],
                                             zoom=4,
                                             threshold=0.3,
                                             colormap=cm.OrRd):
   """
   Overlay fire prediction using fire colormap with a legend on a folium map (USA-only view).
   """
   pred = prediction.squeeze()
   h, w = pred.shape
   lat_range = np.linspace(bbox[0][0], bbox[1][0], h)
   lon_range = np.linspace(bbox[0][1], bbox[1][1], w)


   m = folium.Map(location=map_center, zoom_start=zoom, max_bounds=True)
   m.fit_bounds(bbox)


   #Create color legend
   color_values = np.linspace(0.3, 1.0, 6)
   colors = [f'#{int(r*255):02x}{int(g*255):02x}{int(b*255):02x}'
             for r, g, b, _ in colormap(color_values)]
   legend = bcm.LinearColormap(colors, vmin=0.3, vmax=1.0, caption="Fire Intensity")
   legend.add_to(m)


   #Plot fire points
   for i in range(h):
       for j in range(w):
           value = float(pred[i, j])
           if value > threshold:
               lat = float(lat_range[i])
               lon = float(lon_range[j])
               rgba = colormap(value)
               rgb = tuple(int(c * 255) for c in rgba[:3])
               hex_color = '#{:02x}{:02x}{:02x}'.format(*rgb)


               folium.CircleMarker(
                   location=[lat, lon],
                   radius=3,
                   color=hex_color,
                   fill=True,
                   fill_color=hex_color,
                   fill_opacity=0.9,
                   popup=f"Fire: {value:.2f}"
               ).add_to(m)


   return m


# Assuming predicted_next is from your model
# predicted_next = model.predict(X[-1][np.newaxis, ...])


# Circle Marker Map
m1 = overlay_fire_colormap_points_with_legend(predicted_next[0])
m1.save("usa_fire_map_circle.html")
m1
```
